```python
import torch
import torch.nn as nn

torch.manual_seed(42) #Pytorch own initializer, different starting point, different valley, different final numbers, same destination, different address

X = torch.tensor([[0.,0.], [0.,1.],[1.,0.], [1.,1.]])
y = torch.tensor([[0.], [1.], [1.], [0.]])

model = nn.Sequential(
    nn.Linear(2,2),  #W1, b1 (init included)
    nn.Sigmoid(),   # Our Sigmoid
    nn.Linear(2,1), #W2, b2
    nn.Sigmoid()
)

print(model[0].weight)

loss_fn = nn.MSELoss()      # np.mean((output - y)**2)
optimizer = torch.optim.SGD(model.parameters(), lr=0.5) #Our -==lr * lines

for epoch in range(10000):
    output = model(X)           # FORWARD
    loss = loss_fn(output, y)

    optimizer.zero_grad()   
    loss.backward()     #Our entire backward cell
    optimizer.step()    #the four weight updates

print(model(X).round(decimals=3))
```
